```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
```

```python
def roc_auc():
    return roc_auc_score(y_test, clf.predict_proba(X_test) , multi_class='ovo')
```

```python
df = pd.read_csv("ido&tdo_mol2vec_300.csv" , index_col=0)
df
```

```python
df_class = pd.read_csv("ido_tdo.csv" , index_col=0)
df_class
```

```python
y = df_class.iloc[:,-1]
y
```

```python
X = df.iloc[:, :]
X
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, stratify=y, random_state=42 , test_size=0.2)
```

```python
clf = SVC(kernel='rbf', random_state=42, probability=True)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("mol2vec 300 svm")
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
roc_auc()
```

```python
clf = SVC(kernel='linear' , random_state=42, probability=True)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("mol2vec 300 svm kernel:linear")
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
roc_auc()
```

```python
clf = SVC(kernel='rbf' , class_weight='balanced' , random_state=42, probability=True)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("mol2vec 300 svm kernel:rbf class_weight = blanaced")
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
roc_auc()
```

```python
clf = SVC(kernel='linear', class_weight='balanced' , random_state=42, probability=True)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("mol2vec 300 svm kernel:linear class_weight=balanced")
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
roc_auc()
```
