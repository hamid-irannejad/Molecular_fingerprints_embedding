```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def roc_auc():
    return roc_auc_score(y_test, clf.predict_proba(X_test) , multi_class='ovo')
```

```python
df = pd.read_csv("ido_tdo_chemdist_dataset.csv", index_col=0)
X = df.iloc[:, :]
X
```

```python
df_class = pd.read_csv("ido_tdo.csv" , index_col=0)
df_class
```

```python
y = df_class.iloc[:,-1]
y
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=42)
```

```python
clf = SVC(probability=True,kernel='rbf', random_state=42)
```

```python
# Fit the model on the scaled training data
clf.fit(X_train, y_train)
```

```python
# Predict on the testing set
y_pred = clf.predict(X_test)
```

```python
print("SVM chemdist")
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
roc_auc()
```

```python
clf = SVC(probability=True,kernel='linear', random_state=42)
```

```python
clf.fit(X_train, y_train)
```

```python
# Predict on the testing set
y_pred = clf.predict(X_test)
```

```python
print("SVM chemdist kernel linear")
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
roc_auc()
```
