```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# best model of morgan 2048-bit metrics for internal test
best_model_bit_internal = pd.read_csv("best_model_bit_internal.csv", index_col=0)
best_model_bit_internal
```

```python
fig = sns.heatmap(best_model_bit_internal, annot=True, fmt=".2f" , center=1.1)
fig.figure.savefig('best_model_bit_internal.png', dpi=600 , bbox_inches='tight')
```

```python
# confusion matrix of morgan bit based 2048 for internal test
cm_data = np.array([[30, 8, 3, 1],
                    [5, 21, 0, 8],
                    [6, 0, 7, 3],
                    [7, 1, 3, 49]])

classes = ['AA', 'AI', 'IA', 'II']

plt.figure(figsize=(8, 6))
fig = sns.heatmap(cm_data, annot=True, fmt="d", cmap='Blues', center=1.5, 
                  xticklabels=classes, yticklabels=classes)
fig.figure.savefig('confusion matrix of morgan bit based 2048 for internal test.png', dpi=600, bbox_inches='tight')
plt.show()
```
