```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# best model of kNN-mol2vec 300-dim with scaffold spliting for internal test
best_model_scaffold_mol2vec = pd.read_csv("scaffold_mol2vec_300_knn.csv", index_col=0)
best_model_scaffold_mol2vec
```

```python
fig = sns.heatmap(best_model_scaffold_mol2vec, annot=True, fmt=".2f" , center=1.5)
fig.figure.savefig('scaffold_kNN_mol2vec300.png', dpi=600 , bbox_inches='tight')
```

```python
# confusion matrix of kNN mol2vec 300 for internal test
cm_data = np.array([[28, 2, 4, 1],
                    [10, 27, 1, 1],
                    [4, 0, 4, 3],
                    [22, 6, 2, 37]])

classes = ['AA', 'AI', 'IA', 'II']

plt.figure(figsize=(8, 6))
fig = sns.heatmap(cm_data, annot=True, fmt="d", cmap='Blues', center=1.5, 
                  xticklabels=classes, yticklabels=classes)
fig.figure.savefig('confmat_scaffold_kNN_mol2vec300.png', dpi=600, bbox_inches='tight')
plt.show()
```
