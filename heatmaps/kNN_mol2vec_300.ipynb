```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# best model of kNN-mol2vec 300-dim with scaffold spliting for internal test
best_model_scaffold_mol2vec_internal = pd.read_csv("knn_mol2vec_300.csv", index_col=0)
best_model_scaffold_mol2vec_internal
```

```python
fig = sns.heatmap(best_model_scaffold_mol2vec_internal, annot=True, fmt=".2f" , center=1.5)
fig.figure.savefig('best_model_scaffold_mol2vec_internal.png', dpi=600 , bbox_inches='tight')
```
