```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# best model of morgan 2048-count-based with balanced class weight metrics for internal test
best_model_count_internal_balanced = pd.read_csv("best_model_count_internal_balanced.csv", index_col=0)
best_model_count_internal_balanced
```

```python
fig = sns.heatmap(best_model_count_internal_balanced, annot=True, fmt=".2f" , center=1.2)
fig.figure.savefig('best_model_count_internal_balanced.png', dpi=600 , bbox_inches='tight')
```

```python
# confusion matrix of morgan count based 2048 for internal test with balanced weight
cm_data = np.array([[31, 6, 3, 2],
          [3, 25, 0, 6],
          [5, 0, 7, 4],
          [5, 0, 3, 52]])



classes = ['AA', 'AI', 'IA', 'II']

plt.figure(figsize=(8, 6))
fig = sns.heatmap(cm_data, annot=True, fmt="d", cmap='Blues', center=1.5, 
                  xticklabels=classes, yticklabels=classes)
fig.figure.savefig('confusion matrix of morgan count based 2048 for internal test.png', dpi=600, bbox_inches='tight')
plt.show()
```
