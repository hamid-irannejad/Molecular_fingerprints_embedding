```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# best model of kNN-morgan 1024 with scaffold spliting for internal test
best_model_scaffold_morgan_internal = pd.read_csv("scaffold_knn_morgan_1024.csv", index_col=0)
best_model_scaffold_morgan_internal
```

```python
fig = sns.heatmap(best_model_scaffold_morgan_internal, annot=True, fmt=".2f" , center=1.5)
fig.figure.savefig('scaffold_kNN_morgan1024.png', dpi=600 , bbox_inches='tight')
```

```python
# confusion matrix of kNN morgan 1024 for internal test
cm_data = np.array([[23, 3, 0, 9],
                    [6, 28, 0, 5],
                    [5, 0, 3, 3],
                    [10, 5, 3, 49]])

classes = ['AA', 'AI', 'IA', 'II']

plt.figure(figsize=(8, 6))
fig = sns.heatmap(cm_data, annot=True, fmt="d", cmap='Blues', center=1.5, 
                  xticklabels=classes, yticklabels=classes)
fig.figure.savefig('confmat_scaffold_kNN_morgan1024.png', dpi=600, bbox_inches='tight')
plt.show()
```
