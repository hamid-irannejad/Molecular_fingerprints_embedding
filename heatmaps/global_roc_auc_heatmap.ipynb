```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# heatmap for ROC_AUC score of all models
roc_score = pd.read_csv("roc_score.csv", index_col=0)
roc_score
```

```python
fig = sns.heatmap(roc_score, annot=True, fmt=".2f" , center=1)
fig.figure.savefig('roc_auc.png', dpi=600 , bbox_inches='tight')
```
