```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit.Chem import Descriptors
from rdkit.Chem import rdMolDescriptors
from rdkit import DataStructs
from rdkit.Chem import MACCSkeys
```

```python
df = pd.read_csv("test_data.csv")
df
```

```python
from rdkit.Chem import SaltRemover as sr
remover = sr.SaltRemover()
```

```python
mols = []
for i in range(len(df['smiles'])):
    mol_i = Chem.MolFromSmiles(df['smiles'][i])
    mol_i = remover.StripMol(mol_i, dontRemoveEverything=True)
    mols.append(mol_i)
len(mols)
```

```python
i = 0
l2 = np.zeros((1, 167))
for mol in mols:
    fp2 = MACCSkeys.GenMACCSKeys(mols[i])
    fp2 = fp2.ToList()
    fp2 = np.array(fp2)
    l2 = np.vstack((l2, fp2))
    i += 1
mfp = l2[1:, :]
mfp.shape
```

```python
df_maccskey = pd.DataFrame(mfp)
df_maccskey
```

```python
df_maccskey.to_csv("maccskey_testdata.csv" , index= False)
```
