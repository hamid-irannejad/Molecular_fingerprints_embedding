```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, matthews_corrcoef
```

```python
df = pd.read_csv("morganFP_2048.csv")
df
```

```python
df_class = pd.read_csv("ido_tdo.csv" , index_col=0)
df_class
```

```python
y = df_class.iloc[:,-1]
y
```

```python
X = df.iloc[:, :]
X
```

# **Dimension reduction by t-SNE**

```python
from sklearn.manifold import TSNE
```

```python
X_reduced = TSNE(random_state=42).fit_transform(X)
```

```python
X_reduced
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y = LabelEncoder().fit_transform(y)
```

```python
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from matplotlib.colors import ListedColormap
```

```python
X_scaled = StandardScaler().fit_transform(X_reduced)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, random_state=42 , test_size=0.2)
```

```python
model = SVC()
params = {
    "C": [0.1, 1, 10, 100, 1000],
    "gamma": ["scale", "auto", 1, 0.1, 0.01, 0.001],
    "kernel": ["linear", "rbf", "poly", "sigmoid"]
}
grid = GridSearchCV(model, params, cv=5, n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
clf = SVC(probability=True, C=100, gamma=1, random_state=42)
```

```python
clf.fit(X_train, y_train)
```

```python
ypred = clf.predict(X_test)
```

```python
print("svm morgan reduced by t-sne")
print(classification_report(y_test , ypred))
print(confusion_matrix(y_test , ypred))
print(roc_auc_score(y_test, clf.predict_proba(X_test) , multi_class='ovr'))
print(matthews_corrcoef(y_test, ypred))
```

# **Dimension reduction by PCA**

```python
from sklearn.decomposition import PCA
```

```python
X_reduced = PCA(n_components=2, random_state=42).fit_transform(X)
```

```python
X_reduced
```

```python
X_scaled = StandardScaler().fit_transform(X_reduced)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, random_state=42 , test_size=0.2)
```

```python
model = SVC()
params = {
    "C": [0.1, 1, 10, 100, 1000],
    "gamma": ["scale", "auto", 1, 0.1, 0.01, 0.001],
    "kernel": ["linear", "rbf", "poly", "sigmoid"]
}
grid = GridSearchCV(model, params, cv=5, n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
clf = SVC(C=1000, gamma='auto', random_state=42, probability=True)
```

```python
clf.fit(X_train, y_train)
```

```python
ypred = clf.predict(X_test)
```

```python
print("svm morgan reduced by PCA")
print(classification_report(y_test , ypred))
print(confusion_matrix(y_test , ypred))
print(roc_auc_score(y_test, clf.predict_proba(X_test) , multi_class='ovr'))
print(matthews_corrcoef(y_test, ypred))
```

# **Dimension reduction by pacmap**

```python
import pacmap
```

```python
embedding = pacmap.PaCMAP(random_state=42) 
```

```python
X_reduced = embedding.fit_transform(X)
```

```python
X_scaled = StandardScaler().fit_transform(X_reduced)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, random_state=42 , test_size=0.2)
```

```python
model = SVC()
params = {
    "C": [0.1, 1, 10, 100, 1000],
    "gamma": ["scale", "auto", 1, 0.1, 0.01, 0.001]
}
grid = GridSearchCV(model, params, cv=5, n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
clf = SVC(C=1000, gamma=1, random_state=42, probability=True)
```

```python
clf.fit(X_train, y_train)
```

```python
ypred = clf.predict(X_test)
```

```python
print("svm morgan reduced by pacmap")
print(classification_report(y_test , ypred))
print(confusion_matrix(y_test , ypred))
print(roc_auc_score(y_test, clf.predict_proba(X_test) , multi_class='ovr'))
print(matthews_corrcoef(y_test, ypred))
```

### pacmap result was the best so the figure is get based on this dimension reduction method

```python
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    clf,
    X_test,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, edgecolors="k" , label= y_test)
class_labels = ['Class AA', 'Class AI', 'Class IA' , 'Class II']
handles = [plt.Line2D([0], [0], marker='o', color='w', label=class_labels[i], markerfacecolor=plt.cm.jet(i/len(class_labels)), markersize=10) for i in range(len(class_labels))]
plt.title("classification using Support Vector Machine and pacmap on test set")
plt.legend(handles=handles, title="Classes" , framealpha= 0.2)
plt.axis("tight")
plt.show()
```
