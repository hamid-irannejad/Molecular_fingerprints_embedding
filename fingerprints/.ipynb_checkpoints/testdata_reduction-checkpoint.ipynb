```python
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.read_csv('test_count_morgan_1024.csv')
df1
```

```python
df2 = pd.read_csv('testdata_target.csv')
df2
```

```python
df3 = pd.concat([df1, df2], axis=1)
df3
```

```python
for i in range(len(df3)):
    if df3.iloc[i, -1] == 'II':
        print()
```

```python
df4 = df3.where(df3.iloc[:, -1] == 'II')
df4
```

```python
#Retrieving the index of II class
II_list = list(df4.dropna().index)
```

```python
print(II_list)
```

```python
len(II_list)
```

```python
#Index of reduced II class
new_II = []
i = 0
while i < 74:
    new_II.append(II_list[i])
    i += 2
print(new_II)
print(len(new_II))
```

```python
#Index of classes AA, AI and IA
index_list = []
for i in range(len(df3)):
    if i not in II_list:
        index_list.append(i)
print(index_list)
print(len(index_list))
```

```python
index_list.extend(new_II)
```

```python
len(new_II), len(index_list)
```

```python
final_indices = sorted(index_list)
```

```python
print(final_indices)
```

```python
df5 = df3.iloc[final_indices]
df5
```

```python
#Now, it is better
sns.histplot(df5['ido_tdo'])
```
