```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit.Chem import Descriptors
from rdkit.Chem import rdMolDescriptors
from rdkit import DataStructs
```

```python
df = pd.read_csv("ido&tdo_ChemDist.csv")
df
```

```python
from rdkit.Chem import SaltRemover as sr
remover = sr.SaltRemover()
```

```python
mols = []
for i in range(len(df['smiles'])):
    mol_i = Chem.MolFromSmiles(df['smiles'][i])
    mol_i = remover.StripMol(mol_i, dontRemoveEverything=True)
    mols.append(mol_i)
len(mols)
```

```python

#Morgan Fingerprints:
fpgen2 = AllChem.GetMorganGenerator(fpSize=1024)
i = 0
l2 = np.zeros((1, 1024), dtype='uint8')
for mol in mols:
    fp2 = fpgen2.GetFingerprintAsNumPy(mols[i])
    l2 = np.vstack((l2, fp2))
    i += 1
mfp = l2[1:, :]
mfp.shape
mfp.sum()
```

```python
df2 = pd.DataFrame(mfp)
```

```python
df2.to_csv("morganFP_1024.csv", index=False)
```
