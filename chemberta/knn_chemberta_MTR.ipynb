```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
def roc_auc():
    return roc_auc_score(y_test, clf.predict_proba(X_test) , multi_class='ovo')
```

```python
df = pd.read_csv("chemberta_idotdo_MTR.csv" , index_col=0)
df
```

```python
df_class = pd.read_csv("ido_tdo.csv" , index_col=0)
df_class
```

```python
y = df_class.iloc[:,-1]
y
```

```python
X = df.iloc[:, :]
X
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, stratify=y, random_state=42 , test_size=0.2)
```

```python
# using default values

clf = KNeighborsClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
ypred = clf.predict(X_test)
```

```python
print("knn chemberta mtr")
print(classification_report(y_test , ypred))
print(confusion_matrix(y_test , ypred) , roc_auc())
```

```python
# using default values

clf = KNeighborsClassifier(weights='distance')
```

```python
clf.fit(X_train, y_train)
```

```python
ypred = clf.predict(X_test)
```

```python
print("knn chemberta mtr weight = distance")
print(classification_report(y_test , ypred))
print(confusion_matrix(y_test , ypred) , roc_auc())
```
